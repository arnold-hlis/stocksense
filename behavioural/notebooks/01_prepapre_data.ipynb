```python
import yfinance as yf
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# select or choose the stock

symbol = input("Enter the stock symbol : ")

stock = yf.Ticker(symbol)
print(f"Data fetched for {symbol}")
```

```python
# fetch the news
news_data = stock.news
# prepare news with proper format
stock_news = []
# extract the news information
for news in news_data:
    stock_news.append(
        {
            "title": news["content"]["title"],
            "publisher": news["content"]["provider"]["displayName"],
            "published": news["content"]["pubDate"],
            "summary": news["content"]["summary"],
            "url": news["content"]["canonicalUrl"]["url"],
        }
    )
stock_news = pd.DataFrame(stock_news)
stock_news["published"] = pd.to_datetime(
    stock_news["published"], utc=True
).dt.tz_convert("Asia/Kolkata")
# Sort by oldest to newest for sequential modeling
stock_news = stock_news.sort_values(by="published").reset_index(drop=True)
print("Stock news fetched...")
```

```python
# fetch the fundamental information
stock_fundamentals = stock.info
fundamental_data = {
    "PE Ratio": stock_fundamentals.get("trailingPE"),
    "EPS": stock_fundamentals.get("trailingEps"),
    "Revenue": stock_fundamentals.get("totalRevenue"),
    "Profit Margin": stock_fundamentals.get("profitMargins"),
    "Debt to Equity": stock_fundamentals.get("debtToEquity"),
    "Overall Risk": stock_fundamentals.get("overallRisk"),
    "Regular Market Volume": stock_fundamentals.get("regularMarketVolume"),
    "Marketcap": stock_fundamentals.get("marketCap"),
    "52W Low": stock_fundamentals.get("fiftyTwoWeekLow"),
    "52W High": stock_fundamentals.get("fiftyTwoWeekHigh"),
    "Book Value": stock_fundamentals.get("bookValue"),
}

fund_df = pd.DataFrame(fundamental_data, index=[0]).round(2)
print(f"{symbol} fundamental loaded...")
```
